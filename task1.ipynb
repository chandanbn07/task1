```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/content/API_SP.POP.TOTL_DS2_en_csv_v2_76253.csv"
df = pd.read_csv(file_path, skiprows=4)

# Select relevant columns
year = "2023"
df_filtered = df[["Country Name", year]].dropna()

# Convert population to numeric
df_filtered[year] = pd.to_numeric(df_filtered[year], errors="coerce")

# Bar chart: Top 10 most populated countries
top_10 = df_filtered.nlargest(10, year)

plt.figure(figsize=(12, 6))
plt.bar(top_10["Country Name"], top_10[year], color="skyblue")
plt.xticks(rotation=45, ha="right")
plt.xlabel("Country")
plt.ylabel("Population")
plt.title(f"Top 10 Most Populated Countries in {year}")
plt.show()

# Histogram: Distribution of population sizes
plt.figure(figsize=(10, 6))
plt.hist(df_filtered[year], bins=30, color="orange", edgecolor="black")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.title(f"Distribution of Population Sizes in {year}")
plt.xscale("log")  # Log scale for better visualization
plt.show()
```
